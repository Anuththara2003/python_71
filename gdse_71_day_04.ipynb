def add_corresponding_numbers(list1, list2):
    if len(list1) != len(list2):
        print("Lists must be of the same length.")
        return
    result = []
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])
    return result

def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0:
            total = total + num
            count = count + 1

    if count == 0:
        print("No positive numbers provided.")
        return 0
    else:
        return total / count


print(average_positive(1, -2, 3, 4, -5))
print(average_positive(-1, -2, -3))
print(average_positive(10, 20, 30))
