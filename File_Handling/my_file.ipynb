# **What is JSON?**

* **JSON** stands for **JavaScript Object Notation**.
* It is a **lightweight data format** used for **storing and exchanging data**.
* JSON is **text-based** and **easy for humans to read and write**.
* It is widely used in **web applications** to send data between a server and a client.

---

### **JSON Structure**

JSON is made of **key-value pairs**, similar to Python dictionaries.

#### Example:

```json
{
  "name": "Shasidu",
  "age": 17,
  "is_student": true,
  "courses": ["Python", "Java", "Math"],
  "address": {
    "city": "Colombo",
    "country": "Sri Lanka"
  }
}
```

* `"name"`, `"age"`, `"is_student"` → **keys**
* `"Shasidu"`, `17`, `true` → **values**
* Values can be:

  * String → `"Hello"`
  * Number → `123`
  * Boolean → `true` / `false`
  * Array → `["Python", "Java"]`
  * Object → `{ "key": "value" }`
  * `null` → represents no value

---

### **JSON in Python**

* In Python, JSON is handled using the **`json` module**.
* **Convert Python dictionary to JSON** (serialization):

```python
import json

data = {"name": "Shasidu", "age": 17}
json_string = json.dumps(data)  # Convert dict to JSON string
print(json_string)
```

* **Convert JSON to Python dictionary** (deserialization):

```python
import json

json_string = '{"name": "Shasidu", "age": 17}'
data = json.loads(json_string)  # Convert JSON string to dict
print(data["name"])
```

---

✅ **Summary**:

* JSON = text format for exchanging data.
* Easy to read, write, and parse.
* Works with almost all programming languages, not just JavaScript.





# **What is CSV?**

* **CSV** stands for **Comma-Separated Values**.
* It is a **simple file format** used to **store tabular data** (like a spreadsheet or database table).
* Each line in the file represents a **row**, and **columns are separated by commas** (or sometimes other delimiters like `;`).
* CSV files are **text files**, so they can be opened in **Excel, Notepad, or any text editor**.

---

### **CSV Example**

```
Name,Age,Course
Shasidu,17,Python
Maleesha,18,Java
Nimal,19,Math
```

* **First row** often contains **column headers**.
* **Subsequent rows** contain the data.

---

### **CSV in Python**

Python has a built-in **`csv` module** to work with CSV files.

#### **Reading a CSV file:**

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  # each row is a list
```

#### **Writing to a CSV file:**

```python
import csv

data = [
    ["Name", "Age", "Course"],
    ["Shasidu", 17, "Python"],
    ["Maleesha", 18, "Java"]
]

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

---

### **Key Points**

1. CSV is **plain text**, easy to read and edit.
2. Columns are **separated by commas** (or other delimiters).
3. Widely used for **data import/export** between programs like Excel, databases, and Python.
4. In Python, `csv.reader` → read, `csv.writer` → write.

````python
# Read the input file
with open("input.txt", "r") as infile:
    lines = infile.readlines()

# Write only even-numbered lines to new file
with open("even-line.txt", "w") as outfile:
    for index, line in enumerate(lines, start=1):  # start counting from 1
        if index % 2 == 0:  # even line
            outfile.write(line)
````

````python
# Read file
lines = open("input.txt").readlines()

# Write only even lines
out = open("even-line.txt", "w")

for i in range(len(lines)):
    if (i + 1) % 2 == 0:   # even line number
        out.write(lines[i])

out.close()
````
