```python
# Take three numbers as input
number_1 = float(input("Enter first number: "))
number_2 = float(input("Enter second number: "))
number_3 = float(input("Enter third number: "))

# Display the input numbers
print("Number 1:", number_1)
print("Number 2:", number_2)
print("Number 3:", number_3)

# Initialize variables
largest_one = None
largest_two = None

# Compare to find the largest and second largest
if number_1 > number_2 and number_1 > number_3:
    largest_one = number_1
    if number_2 > number_3:
        largest_two = number_2
    else:
        largest_two = number_3
elif number_2 > number_1 and number_2 > number_3:
    largest_one = number_2
    if number_1 > number_3:
        largest_two = number_1
    else:
        largest_two = number_3
else:
    largest_one = number_3
    if number_1 > number_2:
        largest_two = number_1
    else:
        largest_two = number_2

# Display results
print("Largest number is:", largest_one)
print("Second largest number is:", largest_two)
```

```python
# Structural Pattern Matching
value = int(input("Enter a value: "))

match value:
	case 1:
		print("value is one")
	case 2:
		print("value is two")
	case 3:
		print("value is three")
	case _:
		print("value is other")
```

```python
day = input("Enter your day (mon, tue, wed, thu, fri, sat, sun): ").lower()

match day:
    case "mon":
        print("weekday")
    case "tue":
        print("weekday")
    case "wed":
        print("weekday")
    case "thu":
        print("weekday")
    case "fri":
        print("weekday")
    case "sat":
        print("weekend")
    case "sun":
        print("weekend")
    case _:
        print("Invalid day entered")
```

```python
day = input("Enter your day (mon, tue, wed, thu, fri, sat, sun): ").lower()

weekdays = ["mon", "tue", "wed", "thu", "fri"]
weekends = ["sat", "sun"]

if day in weekdays:
    print("weekday")
elif day in weekends:
    print("weekend")
else:
    print("Invalid day entered")
```

```python
numbers = [10 , 12, 14]

match numbers:
    case [x , y]:
        hypotennuse = (x** 2 + y**2) ** 0.5
        print(hypotennuse)
    case [x , y , z]:
        s = (x + y + z) / 2
        area = (s*(s-x)*(s-y)*(s-z))**0.5
        print("Area of a triangle : ",area)
    case _:
        print("unsupported insput")
```
