```python
my_tuple = {"tennis", "cricket", "football"}
print(my_tuple)
print(type(my_tuple))
print(len(my_tuple))
```

```python
my_tuple = {"tennis", "cricket", "football","tennis"}
print(my_tuple)
print(type(my_tuple))
print(len(my_tuple))
# not supports duplicate values
```

```python
my_tuple = {"tennis", "cricket", "football"}
my_tuple.add("baseball")
print(my_tuple)
print(type(my_tuple))
print(len(my_tuple))
# add new value for sets
```

```python
my_sets = {"tennis", "cricket", "football","tennis"}
my_sets.update(["golf","baseball"])
print(my_sets)
print(type(my_sets))
print(len(my_sets))
# update new values for the sets
```

```python
my_sets = {"tennis", "cricket", "football"}
my_sets.remove("tennis")
print(my_sets)
print(type(my_sets))
print(len(my_sets))
# remove values for the sets
```
