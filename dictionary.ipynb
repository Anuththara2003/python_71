```python
#Dictionaries values can be of any data type
my_dict = {
    "name": "John", #String
    "age": 30,     #Integer
    "height": 5.9, #Float
    "is_student": False, #Boolean
    "courses": ["Math", "Science"], #List
    "address": { #Nested Dictionary
        "street": "123 Main St",
        "city": "New York"
    }
}
print(my_dict)
```

```python
this_dic = {
    "name" : "sugar",
    "weight" : "1kg",
    # "price" : 150.35
}
this_dic.update({"weight" : "10kg","expirey":"2025-01-01"})
print(this_dic)
```

```python
this_dic = {
    "name" : "sugar",
    "weight" : "1kg",
    "price" : 150.35
}

this_dic.pop("price") # remove a key
print(this_dic)

del this_dic["weight"]
print(this_dic)

this_dic.clear()
print(this_dic)
```
