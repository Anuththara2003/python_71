# Python Arithmetic Example: Addition

## Explanation
- 2 + 3 uses the **arithmetic addition operator +** to perform *mathematical addition*.  
- In mathematics, this is equivalent to:  
  \[
  2 + 3 = 5
  \]  
- The result of 2 + 3 is 5.  
- X = 2 + 3 *assigns* the result 5 to the variable X.  
- print(X) displays the value of X, which is 5.  

### Key Points
- The + operator is a *basic arithmetic operator* in Python.  
- Python can perform all standard mathematical operations like addition, subtraction, multiplication, division, exponentiation, and modulus.  
- Variables can store the result of *mathematical expressions* for later use.

````python
X = 2 + 3
print(X)
y = X ** 5
print(y)

````

# Python Assignment Operators Example

## Code
```python
# Basic assignment
x = 2 + 3
print("x =", x)  # Output: x = 5

# Exponentiation assignment
y = x ** 2
print("y =", y)  # Output: y = 25

# Division
z = y / x
print("z =", z)  # Output: z = 5.0

# Addition assignment
x += 4   # Equivalent to: x = x + 4
print("x after += 4:", x)  # Output: 9

# Subtraction assignment
x -= 2   # Equivalent to: x = x - 2
print("x after -= 2:", x)  # Output: 7

# Multiplication assignment
x *= 3   # Equivalent to: x = x * 3
print("x after *= 3:", x)  # Output: 21

# Division assignment
x /= 7   # Equivalent to: x = x / 7
print("x after /= 7:", x)  # Output: 3.0

# Modulus assignment
x %= 2   # Equivalent to: x = x % 2
print("x after %= 2:", x)  # Output: 1.0

# Exponentiation assignment
x **= 3  # Equivalent to: x = x ** 3
print("x after **= 3:", x)  # Output: 1.0

# Floor division assignment
x //= 2  # Equivalent to: x = x // 2
print("x after //= 2:", x)  # Output: 0.0

# Detecting Duplicates in a Tuple using Set in Python

## Code
```python
product_ids = (101, 102, 103, 104, 105, 106, 107, 103, 109)

# Convert tuple to set
Unique_product_ids = set(product_ids)

# Compare lengths
has_duplicate = len(product_ids) != len(Unique_product_ids)

print("Has Duplicates:", has_duplicate)

# Membership Operators in Python

## Membership operators are used to test whether a value or variable exists in a sequence such as lists, tuples, strings, or sets.

### Membership Operators

# in

Returns True if a value is found in the sequence.

fruits = ["apple", "banana", "cherry"]
print("banana" in fruits)  # True
print("orange" in fruits)  # False

# not in

Returns True if a value is not found in the sequence.

fruits = ["apple", "banana", "cherry"]
print("orange" not in fruits)  # True
print("apple" not in fruits)   # False
