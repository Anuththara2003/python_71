## ðŸŸ¦ Constructor in Python

A **constructor** is a special method named `__init__` in a class.
It runs **automatically** when an object is created.

### âœ” What a constructor does:

* Initializes (sets) the object's data
* Prepares the object so it is ready to use
* Assigns values to **instance attributes**

### âœ” Example (corrected version of your code)

```python
class Car:
    wheel = 4   # class attribute

    def __init__(self, brand, color):   # constructor
        self.brand = brand              # instance attribute
        self.color = color              # instance attribute
```

---

## ðŸŸ¦ Instance Attributes

**Instance attributes** are variables that belong to **each object** individually.

They are always written using:

```
self.attribute_name
```

### âœ” Why self?

`self` refers to the **current object**, so the values belong to *that* specific car.

### âœ” Example of creating objects

```python
car1 = Car("Toyota", "Red")
car2 = Car("Nissan", "Blue")

print(car1.brand)   # Toyota
print(car2.color)   # Blue
```

Each object has its **own** brand and color because they are **instance attributes**.

# ðŸŸ¦ Summary

### âœ” Constructor

* Method name: `__init__`

* Initializes object attributes

### âœ” Instance Attributes

* Variables that belong to each object

````python
class Car:
    wheels = 4
    def __init__(self, brand, color): # constructor method
        self.brand = brand # instance attributes
        self.color = color
        
    def driver_to(self, destination):
        return f"The {self.color} {self.brand} is driving to {destination}."
    
    def show_weels_count(self):
        return f"the {self.color} {self.brand} runs on {self.wheels} wheels."

my_car = Car("Toyota", "Red")
print(my_car.driver_to("New York"))
print(my_car.show_weels_count())
````

````python
from math import pi
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
            return pi * (self.radius ** 2)
        
    def circumference(self):
            return 2 * pi * self.radius
    
    def display_info(self):
          print(f"Circle with radius: {self.radius}, Area: {self.area():.2f}, Circumference: {self.circumference():.2f}")


my_circle = Circle(5)
my_circle.display_info()
````
