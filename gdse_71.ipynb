In the code above, `my_list[1:3]` uses Python's list slicing syntax. This expression returns a new list containing elements from index 1 up to, but not including, index 3.

Given `my_list = [10, 20, 30, 40, 50, 60, 70]`, `my_list[1:3]` will return `[20, 30]`.

- The number before the colon (`1`) is the starting index (inclusive).
- The number after the colon (`3`) is the ending index (exclusive).
- So, it selects the elements at index 1 and 2.

```python
my_list = [10,20,30,40,50,60,70]
print(my_list[1:3])  #  print the numbers  between  1 to 5
```

In the code above, `my_list[1:]` uses Python's list slicing syntax. This expression returns a new list containing all elements of `my_list` starting from index 1 (the second element) to the end of the list.

Given `my_list = [10, 20, 30, 40, 50, 60, 70]`, `my_list[1:]` will return `[20, 30, 40, 50, 60, 70]`.

- The number before the colon (`1`) is the starting index (inclusive).
- Omitting the number after the colon means it goes up to the end of the list.

```python
my_list = [10,20,30,40,50,60,70]
print(my_list[1:]) # up to 2nd index
```

In the code above, `my_list[:3]` uses Python's list slicing syntax to select a subset of elements from the list. 

- The expression `my_list[:3]` means "start from the beginning of the list (index 0) and go up to, but not including, index 3."
- So, it returns the elements at index 0, 1, and 2.

Given `my_list = [10, 20, 30, 40, 50, 60, 70]`, `my_list[:3]` will return `[10, 20, 30]`.

This is useful when you want to extract the first few elements from a list without knowing its length.

```python
my_list = [10,20,30,40,50,60,70]
print(my_list[:3]) # back to 2nd index
```

In the code above, `my_list[-3]` uses negative indexing to access elements from the end of the list. In Python, negative indices count from the end of the list, with `-1` referring to the last element, `-2` to the second last, and so on.

Given `my_list = [10, 20, 30, 40, 50, 60, 70]`, `my_list[-3]` will return `50`, which is the third element from the end of the list.

Here's how the negative indices map to the elements:
- `my_list[-1]` → 70
- `my_list[-2]` → 60
- `my_list[-3]` → 50

This feature is useful when you want to access elements relative to the end of a list without needing to know its exact length.

```python
my_list = [10,20,30,40,50,60,70]
print(my_list[-3]) 
```

In the code above, we start with a list `my_list = [10, 8, 9, 15, 20, 18, 26]`. The line `my_list[2:5] = [90, 60, 80]` uses list slicing and assignment to replace a portion of the list.

Here's a step-by-step explanation:

1. **Original List**:  
    `[10, 8, 9, 15, 20, 18, 26]`

2. **Slicing**:  
    `my_list[2:5]` selects the elements at indices 2, 3, and 4, which are `[9, 15, 20]`.

3. **Assignment**:  
    We assign `[90, 60, 80]` to this slice. This means the elements at indices 2, 3, and 4 are replaced by 90, 60, and 80, respectively.

4. **Resulting List**:  
    The updated list becomes `[10, 8, 90, 60, 80, 18, 26]`.

**Key Points:**
- List slicing allows you to select a range of elements.
- Assigning to a slice replaces that range with new values.
- The length of the replacement list must match the length of the slice for a direct replacement.

This technique is useful for updating multiple elements in a list at once.

```python
my_list=[10,8,9,15,20,18,26]
my_list[2:5] = [90, 60, 80]
print(my_list)
```

When calling `my_list.sort()`, Python will raise a `TypeError` because it cannot compare and order elements of different types (e.g., comparing a string to an integer). In Python 3, sorting a list with mixed data types is not allowed unless a custom key function is provided that can handle all types.

**Example Error:**

```python
my_list = ["apple",32, "rat","Grapes","rabbit"]
my_list.sort()
print(my_list)
```

In the code below, we create a list `my_list = ["apple", "32", "rat", "Grapes", "rabbit"]` that contains both strings representing words and a string representing a number (`"32"`). When we call `my_list.sort()`, Python sorts the list in ascending order based on the Unicode code points of the string elements.

**Key Points:**
- All elements in the list are strings, so Python can sort them without error.
- Sorting is case-sensitive: uppercase letters come before lowercase letters in Unicode order.
- Numeric strings (like `"32"`) are sorted as strings, not as numbers, so `"32"` comes before any word starting with a letter.

**Result:**
After sorting, the list becomes: `['32', 'Grapes', 'apple', 'rabbit', 'rat']`.

**Example:**

```python
my_list = ["apple","32", "rat","Grapes","rabbit"]
my_list.sort()
print(my_list)
```

In the code above, we create two lists: `my_list1 = [10, 20, 5, 8]` and `my_list2 = [9, 13, 15]`. - Only the list on which `extend()` is called is modified.
- Printing `my_list1` shows that it is unaffected by the `extend()` operation on `my_list_1`.

```python
my_list1 = [10, 20, 5, 8]
my_list2 = [9, 13, 15]
my_list_1.extend(my_list2)
print(my_list1)
```
