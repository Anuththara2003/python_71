```python
########## Encapsulation ###########

# Default -> public

# Protected -> _ (single underscore)

class Dog:
    def __init__(self,name):
        self._name = name  # protected attribute

dog_1 = Dog("Buddy")
print(dog_1._name)  # Accessible but should be treated as non-public
```

```python
class Person:
    def __init__(self):
        self._name = "John"  # private attribute

class Employee(Person):
    def display_name(self):
        print("Employee Name:", self._name)  # Accessible within subclass

emp = Employee()
emp.display_name()  # Output: Employee Name: John
```

```python
# Private -> __ (double underscore)

class Cat:
    
    def __init__(self,name):
        self.__name = name  # private attribute

cat_1 = Cat("Whiskers")
#print(cat_1.__name)  # Raises AttributeError: 'Cat' object has no attribute '__name' because it's private and name-mangled
```

```python
class StudentRecord:

    def __init__(self,grades):
        self.__grades = grades

    def add_grade(self,grade):
        if self.__validate_grade(grade):
            self.__grades.append(grade)

        else:
            print("Invalid grade")

    def __validate_grade(self,grade):
            return 0 <= grade <= 100 and isinstance(grade, (int))

    def get_average(self,):
            return self.__calculate_average()

    def __calculate_average(self):
            return sum(self.__grades) / len(self.__grades)  

    
student = StudentRecord([85, 90, 78])
student.add_grade(95)
print("Average Grade:", student.get_average())
```
